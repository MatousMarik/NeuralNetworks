# Perceptron


Implement a class for simulating perceptron. The class will have methods for initialization, training and evaluating performance for a perceptron.

    Perceptron(init_weights=None, lr=1.0, max_epochs=1000)
    
**Parameters:**
* `init_weights` (numpy array of floats, default `None`) is the initial value for 
  the extended weight vector, where the last element
  represents the bias.
* `lr` (float, default `1.0`) is the learning rate.
* `max_epochs` (integer, default `1000`) is the maximal number of epochs used when training.

**Attributes:**
* `weights` is extended weight vector of floats, where the last element
  represents the bias.
* `lr` (float) is the learning rate.
* `max_epochs` (integer) is the maximal number of epochs used when training.

**Methods:**
* `predict(X)` predicts class labels for samples in `X`. Input `X` can be
  * vector, i.e., one sample,
  * or a two-dimensional array, where each row is a sample.
  
  Returns
  * either one value 0/1 if the input was a single vector,
  * or vector with values 0/1 with the output of the perceptron 
    for all samples in `X`.
* `partial_fit(X, y, lr=1.0)` perform one epoch perceptron learning algorithm 
  on the training set `X` (two-dimensional numpy array of floats) with 
  the desired outputs `y` (vector of integers 0/1) and learning rate `lr`. Instead of adding or subtracting   an (extended) vector `x`, the function adds or subtracts `lr * x`.
* `fit(X, y, lr=None, max_epochs=None)` trains perceptron using perceptron learning algorithm
  on the training set `X` (two-dimensional numpy array of floats) with 
  the desired outputs `y` (vector of integers 0/1). Returns the number of epochs used in the training 
  (at most `max_epochs`).
* `score(X, y)` returns the mean accuracy on the test data 
  `X` (two-dimensional numpy array of floats) and labels `y` (vector of integers 0/1).

```python
class Perceptron:
    def __init__(self, init_weights=None, lr=1, max_epochs=1000):
        # Perceptron constructor
        self.weights = init_weights
        self.lr = lr
        self.max_epochs = 1000
    
    # learning w.x := p == y -> pass
    #         else p == 0 -> w + lr*x
    #         else p == 1 -> w - lr*x 
    # => w + lr * x * (y - p)
    # p == y => w + 0
    # p == 0 => w + 1*...
    # p == 1 => w - 1*...
        
    def predict(self, X):
        # Compute output of the perceptron.
        # Input X can be
        #  * vector, i.e., one sample
        #  * or a two-dimensional array, where each row is a sample.
        # Returns
        #  * either one value 0/1 if the input was a single vector
        #  * or vector with values 0/1 with the output of the perceptron 
        #    for all samples in X
        if X.ndim == 1:
            X_ = np.append(X, 1)
        else:
            X_ = np.column_stack([X, np.ones(X.shape[0])])
        return (np.dot(X_, self.weights) >= 0).astype(int)
        
    
    def partial_fit(self, X_, y, lr=1.0):
        # perform one epoch perceptron learning algorithm 
        # on the training set `X` (two-dimensional numpy array of floats) with 
        # the desired outputs `y` (vector of integers 0/1) and learning rate `lr`.

        for x, g in zip(X_, y):
            self.weights += (g - (np.dot(x, self.weights) >= 0)) * lr * x
        
            
    def fit(self, X, y, lr=None, max_epochs=None):
        # trains perceptron using perceptron learning algorithm
        # on the training set `X` (two-dimensional numpy array of floats) with 
        # the desired outputs `y` (vector of integers 0/1). If the learning rate `lr == None`, 
        # `self.lr` is used. If `max_epochs == None`, self.max_epochs is used. 
        # Returns the number of epochs used in the training (at most `max_epochs`).
        if lr is None:
            lr = self.lr
        if max_epochs is None:
            max_epochs = self.max_epochs
        X_ = np.column_stack([X, np.ones(X.shape[0])])
        if self.weights is None:
            self.weights = np.random.uniform(-0.3, 0.3, X_.shape[-1])
        for i in range(max_epochs):
            if self.score(X, y) == 1:
                return i
            self.partial_fit(X_, y, lr)
        return max_epochs
    
    def score(self, X, y):
        # returns the mean accuracy on the test data 
        # `X` (two-dimensional numpy array of floats) and 
        # labels `y` (vector of integers 0/1).
        return np.mean(self.predict(X) != y)
```

```python
perc = Perceptron(init_weights=np.ones(3))
print(perc.predict(np.array([1, 2])))
print(perc.predict(np.array([[1,2],[3,4]])))
```

`perc_plot(perc, X, y)` is an auxiliary function that works **for two-dimensional inputs only**. It plots positive and negative samples and the separating hyperplane defined by the current extended weight vector of the pereptron.
  * `perc` is the perceptron.
  * `X` (two-dimensional numpy array of floats) is the array of inputs (each row is one observation), 
    it   must have 2 columns.
  * `y` (vector of integers 0/1) is a vector of desired outputs for observations in `X`.

```python
def perc_plot(perc, X, y):
    # Plot the decision boundary defined by the perceptron perc
    # and the samples - positive samples in red, negative samples in blue color.
    if perc.weights.shape[0] != 3:
        raise Exception("Plotting is possible for 2-dim inputs only")
    plt.plot(X[y == 1, 0],X[y == 1, 1],'+r',label='Positive')
    plt.plot(X[y == 0, 0],X[y == 0, 1],'xb', label='Negative')
    xr = np.array([X[:,0].min(),X[:,1].max()])
    plt.plot(xr,-(perc.weights[0]*xr+perc.weights[2]) / perc.weights[1],'g', label='Separating\nhyperplane')
    plt.legend()
    plt.title("Perceptron decision boundary")
```

Below we illustrate the initial state of the perceptron. It is evident, that the perceptron will classify all training samples in the same way, as they are on the same side of the separating hyperplane.

```python
perc_plot(perc, X, y)               # plot the samples and the separating
                                    # hyperplane of the perceptron
```

Next we train the perceptron and plot its separating hyperplane after learning.

```python
print("The error rate before training:", perc.score(X, y))

it = perc.fit(X, y, 0.5, 100) 

print("The error rate after training:", perc.score(X, y))
print("Used epochs:",it)
```

```python
perc_plot(perc, X, y)
```
